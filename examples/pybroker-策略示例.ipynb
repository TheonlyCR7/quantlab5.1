```python
import sys
sys.path.append('../')
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'
%matplotlib inline
```

```python
from datafeed.dataloader import CSVDataloader
data = CSVDataloader.get_df(['000300.SH', #沪深300
                          '000905.SH', #创业板
                        '159915.SZ'
                         ])
data.dropna()
```

```python
import talib
import pybroker

rsi_20 = pybroker.indicator('rsi_20', lambda data: talib.RSI(data.close, timeperiod=20))
rsi_20(data)
```

```python
def buy_and_hold(ctx):
    if ctx.long_pos():
        return
    ctx.buy_shares = ctx.calc_target_shares(0.3)

from pybroker import Strategy
strategy = Strategy(data, '20100101','20240601')
strategy.add_execution(buy_and_hold, ['000300.SH', '000905.SH','159915.SZ'])
#strategy.add_execution(short_high, ['159915.SZ'])
result = strategy.backtest()
result.orders
```

```python
import matplotlib.pyplot as plt

chart = plt.subplot2grid((3, 2), (0, 0), rowspan=3, colspan=2)
chart.plot(result.portfolio.index, result.portfolio['market_value'])
```
