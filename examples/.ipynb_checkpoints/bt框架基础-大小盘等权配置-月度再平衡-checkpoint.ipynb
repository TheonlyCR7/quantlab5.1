```python
import bt
import sys
sys.path.append('../../')
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'
%matplotlib inline
```

```python
from datafeed.dataloader import CSVDataloader
data = CSVDataloader.get(['000300.SH','159915.SZ'])
data.dropna(inplace=True)
(data.pct_change()+1).cumprod().plot()
```

```python
data.pct_change().corr()
```

```python
import bt
s = bt.Strategy('大小盘-等权-月度再平衡', [bt.algos.RunMonthly(),
                       bt.algos.SelectAll(),
                       bt.algos.WeighEqually(),
                       bt.algos.Rebalance()])
test = bt.Backtest(s, data)
stras = [test]
```

```python
for bench in ['000300.SH','159915.SZ']:
    data = CSVDataloader.get([bench])
    s = bt.Strategy(bench, [bt.algos.RunOnce(),
                       bt.algos.SelectAll(),
                       bt.algos.WeighEqually(),
                       bt.algos.Rebalance()])
    stra = bt.Backtest(s, data)
    stras.append(stra)
    
res = bt.run(*stras)
```

```python
res.plot()
```
