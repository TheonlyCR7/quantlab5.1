```python
import bt
import sys
sys.path.append('../')
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'
%matplotlib inline
```

```python
from datafeed.dataloader import CSVDataloader
data = CSVDataloader.get(['000300.SH', #300
                          '159915.SZ', #
                          'HSI', #
                          '^NDX',# 100
                          'GDAXI',#DAX
                          'N225',#225
                          'AU.NH',#
                          'SC.NH',# 
                          '000012.SH',#
                         ])
data.dropna(inplace=True)
(data.pct_change()+1).cumprod().plot()
```

```python
(data.pct_change()+1).corr()
```

```python
import bt
from bt_algos_extend import SelectTopK
signal = data.pct_change(20)
#signal.dropna(inplace=True)
all = []
for K in [1,2,3,4,5]:
    s = bt.Strategy('---K={}'.format(K), [
                           SelectTopK(signal,K),
                           bt.algos.WeighEqually(),
                           bt.algos.Rebalance()])
    all.append(s)
stras = [bt.Backtest(s, data) for s in all]
```

```python
for bench in ['000300.SH']:
    data = CSVDataloader.get([bench])
    s = bt.Strategy(bench, [bt.algos.RunOnce(),
                       bt.algos.SelectAll(),
                       bt.algos.WeighEqually(),
                       bt.algos.Rebalance()])
    stra = bt.Backtest(s, data)
    stras.append(stra)
    
res = bt.run(*stras)
```
